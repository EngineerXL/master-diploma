# Voxelized LiDAR Point Cloud GIF Animation

This notebook demonstrates how to create a GIF animation of consecutive voxelized LiDAR point clouds using matplotlib.

```python
# Import necessary libraries
from src.dataset.utils import LidarOdometryWrapper
from src.lidar_odometry.voxelize import voxelize
from src.lidar_odometry.cloud_processing import remove_distant_points
```

```python
# Initialize the LidarOdometryWrapper
# This automatically loads the dataset and creates a rotation matrix around z-axis
wrapper = LidarOdometryWrapper("boreas-2021-05-06-13-19", truck_initial_x=-7.5, truck_overtake_vx=15, simulate_truck=True)

INDEX_START = 7007  # Starting index for the sequence
NUM_FRAMES = 10  # Number of consecutive LiDAR clouds to show
MAX_DIST = 50.0  # Maximum distance filter
VOXEL_SIZE = 1.0  # Voxel size for voxelization

input_clouds = []
frame_indices = list(range(INDEX_START, INDEX_START + NUM_FRAMES))
wrapper.set_first_frame(frame_indices[0])

for i, idx in enumerate(frame_indices):
    # Get the LiDAR point cloud for this frame
    ts, sample_points, velocities = wrapper.get_frame(idx)

    # Filter distant points
    filtered_points = remove_distant_points(sample_points, MAX_DIST)

    # Voxelize the point cloud
    voxelized_points = voxelize(filtered_points, VOXEL_SIZE)

    input_clouds.append(voxelized_points)
```

```python
from src.dataset.visualization import plot_lidar_clouds_animation

gif_path = "../pictures/voxelized_lidar_truck_overtake_animation.gif"
plot_lidar_clouds_animation(input_clouds, gif_path=gif_path)
```
