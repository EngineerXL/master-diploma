# Voxelized LiDAR Point Cloud GIF Animation

This notebook demonstrates how to create a GIF animation of consecutive voxelized LiDAR point clouds using matplotlib.

```python
# Import necessary libraries
from src.dataset.pipeline import LidarOdometryPipeline
from src.dataset.visualization import plot_lidar_clouds_animation
```

```python
rides = {
    "demo": {
        "ride_id": "boreas-2021-05-06-13-19",
        "ride_frame_st": 4200,
        "ride_frame_en": 4300,
    },
    "suburbs": {
        "ride_id": "boreas-2025-01-08-11-22",
        "ride_frame_st": 1100,
        "ride_frame_en": 1200,
    },
    "skyway": {
        "ride_id": "boreas-2024-12-04-12-08",
        "ride_frame_st": 900,
        "ride_frame_en": 1000,
    },
    "tunnel": {
        "ride_id": "boreas-2024-12-04-14-50",
        "ride_frame_st": 1000,
        "ride_frame_en": 1100,
    },
    "industrial": {
        "ride_id": "boreas-2024-12-23-16-27",
        "ride_frame_st": 900,
        "ride_frame_en": 1000,
    },
    "farm": {
        "ride_id": "boreas-2025-07-18-15-30",
        "ride_frame_st": 400,
        "ride_frame_en": 500,
    },
}
```

### Pictures

```python
for label, ride_segment in rides.items():
    plot_lidar_clouds_animation(
        LidarOdometryPipeline(
            {
                "wrapper_settings": {
                    "truck_initial_x": -5,
                    "truck_overtake_vx": 1,
                    "simulate_truck": False,
                },
                "ride_info": ride_segment,
                "actor_settings": None,
                "config_name": "Demo",
            }
        ).visualize(num_frames=1, start_frame_offset = 50),
        img_path=f"../pictures/rides/{label}_frame.png",
    )
```

```python
for label, ride_segment in rides.items():
    plot_lidar_clouds_animation(
        LidarOdometryPipeline(
            {
                "wrapper_settings": {
                    "truck_initial_x": -5,
                    "truck_overtake_vx": 1,
                    "simulate_truck": True,
                },
                "ride_info": ride_segment,
                "actor_settings": None,
                "config_name": "Demo",
            }
        ).visualize(num_frames=1, start_frame_offset=50),
        img_path=f"../pictures/rides/{label}_truck_overtake_frame.png",
    )
```

### Animations

```python
for label, ride_segment in rides.items():
    plot_lidar_clouds_animation(
        LidarOdometryPipeline(
            {
                "wrapper_settings": {
                    "truck_initial_x": -5,
                    "truck_overtake_vx": 1,
                    "simulate_truck": False,
                },
                "ride_info": ride_segment,
                "actor_settings": None,
                "config_name": "Demo",
            }
        ).visualize(),
        frame_duration_ms=1000,
        gif_path=f"../animations/rides/{label}_animation.gif",
    )
```

```python
for label, ride_segment in rides.items():
    plot_lidar_clouds_animation(
        LidarOdometryPipeline(
            {
                "wrapper_settings": {
                    "truck_initial_x": -5,
                    "truck_overtake_vx": 1,
                    "simulate_truck": True,
                },
                "ride_info": ride_segment,
                "actor_settings": None,
                "config_name": "Demo",
            }
        ).visualize(),
        frame_duration_ms=1000,
        gif_path=f"../animations/rides/{label}_truck_overtake_animation.gif",
    )
```
