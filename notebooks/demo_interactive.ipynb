# LiDAR Point Cloud Visualization with LidarOdometryWrapper

This notebook demonstrates how to use the `LidarOdometryWrapper` class to load a segment from pyboreas and visualize the LiDAR point cloud using plotly.

```python
import os
import sys

module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
# Import necessary libraries
from src.dataset.utils import LidarOdometryWrapper
from src.lidar_odometry.cloud_processing import remove_distant_points
from src.lidar_odometry.voxelize import voxelize
```

```python
# Initialize the LidarOdometryWrapper
# This automatically loads the dataset and creates a rotation matrix around z-axis
wrapper = LidarOdometryWrapper("boreas-2021-05-06-13-19")

INDEX = 6000  # Starting index for the sequence
MAX_DIST = 50.0  # Maximum distance filter
VOXEL_SIZE = 1.0  # Voxel size for voxelization

# Use the wrapper's get_lidar_point_cloud method which handles rotation automatically
ts, sample_points, velocities = wrapper.get_frame(INDEX)
filtered_points = remove_distant_points(sample_points, MAX_DIST)
voxelized_points = voxelize(filtered_points, VOXEL_SIZE)
```

## Voxelized Frame

```python
from src.dataset.visualization import plot_lidar_clouds_interactive

fig = plot_lidar_clouds_interactive(voxelized_points, title = "Voxelized LiDAR Point Cloud Visualization")
# Show the plot
fig.show()
```
