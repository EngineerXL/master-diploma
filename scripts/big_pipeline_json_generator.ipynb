```python
from src.dataset.utils import LidarOdometryWrapper
import json
```

```python
ride_ids = [
    "boreas-2024-12-04-12-34",
    "boreas-2024-12-04-15-19",
    "boreas-2024-12-23-16-44",
    "boreas-2025-02-15-17-19",
    "boreas-2024-12-04-12-08",
    "boreas-2024-12-04-14-50",
    "boreas-2024-12-23-16-27",
    "boreas-2025-01-08-11-22",
    "boreas-2025-07-18-15-30",
    "boreas-2025-08-13-09-01",
]

ride_segments = []
RIDE_LEN_FRAMES = 150

for ride_id in ride_ids:
    wrapper = LidarOdometryWrapper(ride_id)
    frames = wrapper.get_frames_count()
    segments = frames // RIDE_LEN_FRAMES
    for i in range(segments):
        ride_segments.append(
            {
                "ride_id": ride_id,
                "ride_frame_st": i * RIDE_LEN_FRAMES,
                "ride_frame_en": (i + 1) * RIDE_LEN_FRAMES,
            }
        )
print(len(ride_segments))
```

```python
with open("resources/small_pipeline.json", "r") as file:
    data = json.load(file)  # Use json.load() for files
```

```python
data["ride_segments"] = ride_segments
with open("resources/big_pipeline.json", "w") as file:
    json.dump(data, file)
print(f"Create big pipeline with {len(ride_segments)} ride segments!")
```
